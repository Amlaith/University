```python
from random import random
from math import ceil
from time import time
import re


def mergesort(data):
    start_time = time()
    'Обертка, открывает файл и вызывает рекурсивную функцию сортировки'
    with open('mergesort_out.txt', 'w', encoding='utf-8') as output: # Открывается output файл
        
        log = [f'Исходные данные:\n {data}\n\n'] # Импровизированная визуализация
        
        if type(data) == str: # Если строка - разбиваем на список слов
            data = list(map(lambda x: x.lower(), re.sub(r'[«».!?—:,;`]', ' ', data).split()))
            
        log.append(f'Подготовленные данные:\n {data}\n\n') # Импровизированная визуализация
        
        log.append('Процесс сортировки: \n') # --//--   
        
        result, inner_log = _mergesort(data, [], 1, len(data), 0) # Вызов сортирующей функции
        log += inner_log
        
        output.write(' '.join(result) + '\r\n') # Отсортированный список записывается в файл
        
        log.append(f'Время выполнения: {time() - start_time} сек.')
        
        for line in log:
            output.write(line)
        
        
        
    
def _mergesort(data, log, node_size, length, comp):
    'Собственно сортировка'
    double_ns = node_size * 2
    
    for i in range(0, length, double_ns): # Проходим по списку кусками нужного размера
        a, b = 0, node_size # Счетчики двух кусочков, сливаемых в один
        a_lim, b_lim = node_size, double_ns # Когда нужно их остановить
        if length - i < double_ns: # Место остановки сдвигается, если элементов не хватает для ровных кусочков
            rem = length - i
            b_lim -= double_ns - rem
            if rem < node_size:
                a_lim = rem 
        
        temp = [] # Временный список, хранит результат объединения двух текущих кусков...
        while a < a_lim or b < b_lim: # ...Постепенно заполняется меньшими из 2х элементов в началах сливаемых кусков...
            if a >= a_lim:
                temp.append(data[i+b])
                b += 1
                continue
            if b >= b_lim:
                temp.append(data[i+a])
                a += 1
                continue
            left, right = data[i+a], data[i+b]
            if left < right:
                temp.append(left)
                a += 1
                comp +=1
            else:
                temp.append(right)
                b += 1
                comp +=1
        data[i:i+double_ns] = temp # ...И занимает свое место в основном списке
        
    cur_log = data[:] # Этот блок записывает в файл текущее состояние списка, разделяя палочками рабочие блоки 
    for i in range(ceil(length/double_ns)):
        cur_log.insert(i*(double_ns+1), '|')
    log.append(' '.join(map(str, cur_log)) + ' |\r\n')
    
    if node_size < length/2: # Если объединение не завершено - переходим на следующую итерацию
        return _mergesort(data, log, double_ns, length, comp)
    
    log.append(f'\nКоличество сравнений: {comp}\n')
    return data, log

    
with open('mergesort_in.txt', encoding='utf-8') as f:
    mergesort(f.read())
```

```python
def bi_search(item, data):
    'Обертка поиска'
    if not data or item > data[-1] or item < data[0]:  
        return -1 # Если список пустой или искомый элемент не в диапазоне между 1м и последним элементами
    return _bi_search(item, data, 0, len(data) - 1)
    

def _bi_search(item, data, left, right):
    'Собственно поиск'
    if data[left] == item: # Если элемент на краю текущего отрезка
        return left
    if data[right] == item:
        return right
    
    if (right - left) < 2: # Если отрезок схлопнулся, и элемент не найден
        return -1
    
    center = (right + left) // 2 # Если элемент внутри отрезка
    if data[center] == item: # Если в центре
        return center
    if data[center] > item:
        return _bi_search(item, data, left, center) # Если левее центра
    else:
        return _bi_search(item, data, center, right) # Если правее центра
    

with open('mergesort_out.txt', encoding='utf-8') as f:
    print(bi_search('элемент', f.readline().split()))
```
