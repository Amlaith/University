#### Спираль филлотаксиса

```python
from math import sin, cos, pi, sqrt
from tkinter import *
```

```python
COLORS_DCT = {
    'Розовый': '#F4CFDF',
    'Голубой': '#B6D8F2',
    'Желтый': '#F7F6CF',
    'Зеленый': '#D5E4C3',
    'Фиолетовый': '#C0ADC5',
    'Белый': '#fcfcfc',
    'Серый': '#DAD5D6',
    'Черный': '#101010',
    } 

COLORS_LIST = ['Розовый', 'Голубой', 'Желтый', 'Зеленый', 'Фиолетовый', 'Белый', 'Серый', 'Черный', ]

ANGLES_DCT = {
    '137.5077778 (Угол филлотаксиса)': 137.5077778,
    '135': 135,
    '136': 136,
    '137': 137,
    '138': 138,
    '139': 139,
    '140': 140,
}

ANGLES_LIST = ['137.5077778 (Угол филлотаксиса)', '135', '136', '137', '138', '139', '140', ]
```

```python
def init_variables():
    global color_chosen
    global ask_angle
    global ask_color
    
    ask_angle = StringVar(root)
    ask_color = StringVar(root)
    
    color_chosen = False
    ask_color.set(COLORS_LIST[0])
    ask_angle.set(ANGLES_LIST[0])
    

def set_variables():
    global color_chosen
    global ask_angle
    global ask_color
    
    color_chosen = False
    ask_color.set(COLORS_LIST[0])
    ask_angle.set(ANGLES_LIST[0])
```

```python
# Класс точек, выстраивающихся в спираль

class Dot:
    def __init__(self, centre, angle, dist, fill_color, outline_color):
        self.fill_color = fill_color
        self.outline_color = outline_color
        self.centre = centre
        self.angle = angle
        self.dist = dist
        self.radius = 6
        self.x, self.y = Dot.p2c(
            self.centre,
            self.angle,
            self.dist,
            )
        self.id = c.create_oval(
            self.x,
            self.y,
            self.x + self.radius * 2,
            self.y + self.radius * 2,
            fill=Dot.h2rgb(self.fill_color),
            outline=self.outline_color,
            tag='dot',
            )

        
    def h2rgb(hue):
        def get_x(h):
            return 1 - abs((h / 60) % 2 - 1) 

        r = g = b = 0

        if hue < 60:
            r = 1
            g = get_x(hue)
        elif hue < 120:
            r = get_x(hue)
            g = 1
        elif hue < 180:
            b = get_x(hue)
            g = 1
        elif hue < 240:
            g = get_x(hue)
            b = 1
        elif hue < 300:
            r = get_x(hue)
            b = 1
        else:
            b = get_x(hue)
            r = 1
        r = int(r * 255)
        g = int(g * 255)
        b = int(b * 255)

        return f'#{r:02x}{g:02x}{b:02x}'
    

    def p2c(centre, angle, dist):
        '''Перевод из полярной системы координат в прямоугольную'''
        x = cos(angle * pi / 180) * dist + centre[0] 
        y = sin(angle * pi / 180) * dist + centre[1]
        return x, y


    def refill(self, refill_speed):
        '''Перекрашивание точки'''
        c.itemconfig(self.id, fill=Dot.h2rgb(self.fill_color))
        self.fill_color = (self.fill_color + refill_speed) % 360


    def rotate(self, spin_speed):
        '''Вращение точки'''
        new_angle = self.angle + spin_speed
        x, y = Dot.p2c(self.centre, new_angle, self.dist)        
        c.move(self.id, (x - self.x), (y - self.y))
        self.x = x
        self.y = y
        self.angle = new_angle
```

```python
# Функция выхода из приложения

def exit_app():
    root.destroy()
```

```python
# Главная функция, рисует 680 точек нужного цвета и в нужном месте, вызывает методы движения и перекраски у существующих точек

def draw():
    global a, j
    
    if j < 680:
        dots.append(
            Dot(
                (193, 193),
                a,
                sqrt(180*j/pi),
                (j*REFILL_SPEED/pi)%360,
                BACKGROUND,
            )
        ) # Новая точка добавляется в список
    
    a += (ANGLE + SPIN_SPEED) # Считается угол появления следующей точки
    a %= 360 
    j += 1 # 
    
    for dot in dots: # раз в два сдвига каждая точка меняет цвет
        if j % 2 == 0:
            dot.refill(REFILL_SPEED)
        dot.rotate(SPIN_SPEED)
    
    c.after(25, draw)
```

```python
def start_screen():
    # Начальный экран выбора цвета и угла
    
    global ask_color
    global ask_angle
    global color_chosen
    
    
    # Удаление всех предыдущих виджетов
    
    for widget in root.winfo_children():
        widget.destroy()

    
    root.geometry('400x470')
    color_chosen = False
    
    # Меню выбора цвета фона с лейблом    
    
#     try:
#         ask_color
#     except NameError:
#         print('ne')
#         ask_color = StringVar(root)
#         ask_color.set(COLORS_LIST[0])
    
    
    BACKGROUND=COLORS_DCT[ask_color.get()]
    
    
    ask_color_label = Label(
        root,
        padx=5,
        pady=2,
        fg='#101010',
        bg=BACKGROUND,
        text='Выберите цвет, который вам больше нравится: ',
        font=("Helvetica", 10),
        )
        
    
    color_menu = OptionMenu(
        root,
        ask_color,
        *COLORS_LIST,
        )

    
    ask_color_label.place(relx=0.5, rely=0.36, anchor=CENTER)
    color_menu.place(relx=0.5, rely=0.42, anchor=CENTER)
    
    
    # Меню выбора угла с лейблом
    
    ask_angle_label = Label(
        root,
        padx=5,
        pady=2,
        fg='#101010',
        bg=BACKGROUND,
        text='Выберите угол появления точек спирали: ',
        font=("Helvetica", 10),
        )
#     ask_angle.set(ANGLES_LIST[0])
    angle_menu = OptionMenu(
        root,
        ask_angle,
        *ANGLES_LIST,
        )
#     angle_menu.config(bg=BACKGROUND)

    
    ask_angle_label.place(relx=0.5, rely=0.54, anchor=CENTER)
    angle_menu.place(relx=0.5, rely=0.60, anchor=CENTER)

    
    # Кнопка подтверждения выбора и перехода к экрану спирали
    
    submit = Button(
        root,
        command=start_spiral,
        fg='#101010',
        text='Продолжить',
    
    
    submit.place(relx=0.5, rely=0.83, anchor=CENTER)
   
    
    # Кнопка выхода из приложения    
    
    exit_button = Button(
        root,
        text='Выйти',
        command=exit_app,
        )
    
    
    exit_button.place(relx=0.5, rely=0.93, anchor=CENTER)
    

    # Предпросмотр выбираемого цвета 
    def preview_bg():
        global BACKGROUND
        global FOREGROUND
    
        if not color_chosen:
            BACKGROUND=COLORS_DCT[ask_color.get()]
            FOREGROUND = '#000000' if BACKGROUND != '#101010' else '#e3e3e3'
            root.configure(bg=BACKGROUND)
            ask_color_label.configure(bg=BACKGROUND, fg=FOREGROUND)
            ask_angle_label.configure(bg=BACKGROUND, fg=FOREGROUND)
        
            root.after(20, preview_bg)
      
    preview_bg()
```

```python
# Переход с начального экрана на спираль
def start_spiral():
    global a
    global j
    global dots
    global c
    global color_chosen
    global ANGLE
    global REFILL_SPEED
    
    
    color_chosen = True
    BACKGROUND = COLORS_DCT[ask_color.get()]
    ANGLE = ANGLES_DCT[ask_angle.get()]

    
    # Виджеты выбора цвета удаляются, появляется канвас
    for widget in root.winfo_children():
        widget.destroy()
    
    
    c = Canvas(
        root,
        width=400,
        height=470,
        bg=BACKGROUND,
    )
    c.pack()


    # Слайдер регулировки скорости вращения с лейблом. Значение 10 раз в секунду записывается в SPIN_SPEED

    spin = Scale(root,
                 from_=0,
                 to=5,
#                  bd=6,
                 orient=HORIZONTAL,
                )
    spin.set(2)
    spin.config(troughcolor=BACKGROUND, length=112)
    spin.place(relx=0.5, rely=0.9, anchor=CENTER)


    def get_spin_speed():
        global SPIN_SPEED
        SPIN_SPEED = spin.get()
        c.after(100, get_spin_speed)



    get_spin_speed()


    spin_label = Label(
        root,
        text='Скорость вращения',
    )
    spin_label.config(width=16)
    spin_label.place(relx=0.5, rely=0.96, anchor=CENTER)


    # Скорость смены цвета

    REFILL_SPEED = 6
    
    
    # Кнопка возврата к начальному экрану    
    
    back_button = Button(
        root,
        text='Назад',
        command=start_screen,
        )
    
    
    back_button.place(relx=0.1, rely=0.93, anchor=CENTER)

    
    # Кнопка выхода из приложения    
    
    exit_button = Button(
        root,
        text='Выйти',
        command=exit_app,
        )
    
    
    exit_button.place(relx=0.9, rely=0.93, anchor=CENTER)
    
    
    # Собственно начало спирали
    
    a = j = 1 # a - угол поворота, j - счетчик точек (и, по совместительству, времени)
    dots = [] # Хранилище всех точек
    
    
    draw()
```

```python
root = Tk()
root.title('Спираль филлотаксиса')
# root.iconbitmap('sunflower.ico')

init_variables()


start_screen()

root.mainloop()
```
