```python
import pygame
from math import sin, cos, pi
from random import randint 


# initialize the pygame module
pygame.init()


pygame.display.set_caption('Domashneye zadanie')

# create a surface on screen that has the size of 240 x 180
screen = pygame.display.set_mode((1920, 1080))

# define a variable to control the main loop
running = True
clock = pygame.time.Clock()

def p2c(centre, angle, dist):
    x = cos(angle*pi/180) * dist + centre[0]
    y = sin(angle*pi/180) * dist + centre[1]
    return int(x), int(y)


# main loop
while running:
    # limit the while loop to a max of 10 times per second
    clock.tick(10)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill((188, 214, 255))

    
    pygame.draw.circle(screen, (0, 0, 0), [960, 540],  3)
    
#     a = pygame.draw.circle(screen, (150, 150, 255), p2c((120, 160), 0, 20),  5)
#     b = pygame.draw.circle(screen, (255, 150, 150), p2c((120, 160), 0, 40),  5)
#     c = pygame.draw.circle(screen, (150, 255, 150), p2c((120, 160), 0, 60),  5)
    class Dot:
        def __init__(self, canvas, centre, angle, layer, radius, color):
            pygame.draw.circle(
                canvas,
                color,
                p2c(centre, angle, layer * (radius * 2 + 1)),
                radius                
            )
      
    
#     d1 = Dot(screen, (120, 160), 90, 2, 5, (255, 222, 209))
    
    class Circle:
        def __init__(self, canvas, centre, layer, dot_radius, color):
            dots_num = layer * 7
            dots = [Dot(canvas, centre, 360/dots_num*i, layer, dot_radius, color) for i in range(dots_num)]
            
#     circle1 = Circle(screen, (120, 160), 2, 5, (245, 234, 255))
    
    
    class Carpet:
        def __init__(self, canvas, centre, layers, dot_radius):
            circles = [Circle(canvas, centre, i, dot_radius, (randint(1, 255), randint(1, 255), randint(1, 255))) for i in range(0, layers+1)]
    
    
    carpet = Carpet(screen, (960, 540), 100, 4)
    
    
    pygame.display.flip()
    
# print(a, b, c)

pygame.quit()
```
